<h1> Splitting data  </h1>

```python
x = data.drop(['csMPa'],axis = 1)
y = data['csMPa']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.35,random_state = 42)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

<h1> Linear Regression </h1>


```python
reg = LinearRegression()
reg.fit(x_train,y_train)
predict = reg.predict(x_test)
predict
```

```python
print("Accuracy on training data :",round(reg.score(x_train,y_train)*100,2),'%')
print("Accuracy on Testing data : ",round(reg.score(x_test,y_test)*100,2),'%')
acc_DT=r2_score(y_test, predict)
print('Accuracy DT: ',acc_DT)
print('MSE: ',mean_squared_error(y_test, predict))
L_results = pd.DataFrame({'Method':['Linear Regression'], 'accuracy': acc_DT},index={'1'})
```

<h1> Ridge Regression </h1>

```python
reg = Ridge(alpha = 0.5)
reg.fit(x_train,y_train)
predict = reg.predict(x_test)
predict
```

```python
print("Accuracy on training data :",round(reg.score(x_train,y_train)*100,2),'%')
print("Accuracy on Testing data : ",round(reg.score(x_test,y_test)*100,2),'%')
acc_DT=r2_score(y_test, predict)
print('Accuracy DT: ',acc_DT)
print('MSE: ',mean_squared_error(y_test, predict))
R_results = pd.DataFrame({'Method':['Ridge Regression'], 'accuracy': acc_DT})
```

<h1> Lasso Regression </h1>

```python
reg = Lasso()
reg.fit(x_train,y_train)
predict = reg.predict(x_test)
predict
```

```python
print("Accuracy on training data :",round(reg.score(x_train,y_train)*100,2),'%')
print("Accuracy on Testing data : ",round(reg.score(x_test,y_test)*100,2),'%')
acc_DT=r2_score(y_test, predict)
print('Accuracy DT: ',acc_DT)
print('MSE: ',mean_squared_error(y_test, predict))
Lo_results = pd.DataFrame({'Method':['Lasso Regression'], 'accuracy': acc_DT})
```

<h1> SVR </h1>

```python
reg =SVR()
reg.fit(x_train,y_train)
predict = reg.predict(x_test)
predict
```

```python
print("Accuracy on training data :",round(reg.score(x_train,y_train)*100,2),'%')
print("Accuracy on Testing data : ",round(reg.score(x_test,y_test)*100,2),'%')
acc_DT=r2_score(y_test, predict)
print('Accuracy DT: ',acc_DT)
print('MSE: ',mean_squared_error(y_test, predict))
svr_results = pd.DataFrame({'Method':['SVR Regression'], 'accuracy': acc_DT})
```

<h1> RANDOM FOREST REGRESSOR </h1>

```python
reg = RandomForestRegressor()
reg.fit(x_train,y_train)
predict = reg.predict(x_test)
predict
```

```python
print("Accuracy on training data :",round(reg.score(x_train,y_train)*100,2),'%')
print("Accuracy on Testing data : ",round(reg.score(x_test,y_test)*100,2),'%')
acc_DT=r2_score(y_test, predict)
print('Accuracy DT: ',acc_DT)
print('MSE: ',mean_squared_error(y_test, predict))
rfa_results = pd.DataFrame({'Method':['Random Forest Regression'], 'accuracy': acc_DT})
```
