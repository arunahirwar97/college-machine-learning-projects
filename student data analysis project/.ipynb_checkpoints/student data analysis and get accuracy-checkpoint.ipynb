#  Top 10 states provide computers in primary class student with every year

```python
# print top 10 state provide computers for primary class with every year 
top_10 = data.groupby(['State_UT','year'])['Primary_Only'].sum().reset_index().sort_values(by = 'Primary_Only',ascending = False)
top_10 =top_10.head(30)
print(top_10)
```

```python
#  Show data by bar chart
plt.figure(figsize = (20,10))
sns.catplot('State_UT','Primary_Only','year',kind = 'bar',data = top_10,aspect = 1.8,height = 8)
plt.xticks(rotation = 90)
plt.show()
```

```python
x = data.drop(['All Schools','year','State_UT'],axis = 1)
y = data['All Schools']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.35)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
reg = RandomForestRegressor(n_estimators = 10)
```

```python
reg.fit(x_train,y_train)
```

```python
print("Accuracy for test data : ",reg.score(x_test,y_test))
```

```python
print('Acccuracy For Training data : ',reg.score(x_train,y_train))
```
