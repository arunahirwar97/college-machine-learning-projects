<h1> Asked many questions so gives answer </h1>

<h1> Question No. 1 : Print top 10 Course name which subscribers is heighest . </h1>

```python
# Print top ten course which is sunscribers is heighest 
grp = data.groupby('course_title')['num_subscribers'].sum().reset_index().sort_values(by = 'num_subscribers',ascending = False)
grp.head(10)
```

```python
plt.figure(figsize =(20,10))
sns.catplot('course_title','num_subscribers',data=grp.head(10),kind='bar',aspect=3,height=8,palette='coolwarm')
plt.title('TOP 10 COURSE NAME WHICH IS HEIGHEST SUBSCRIBERS',size=35,color = 'Green')
plt.xticks(size=25,rotation = 90)
plt.yticks(size=25)
plt.ylabel('SUBSCRIBERS',size=15)
plt.xlabel('COURSE TITLE',size=15)
plt.show()
```
