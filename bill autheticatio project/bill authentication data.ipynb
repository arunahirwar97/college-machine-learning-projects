```python
x = data.iloc[:,0:4].values
y = data.iloc[:,4:5].values
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.30,random_state = 42)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
aa = LogisticRegression()
aa.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("For Train data Prediction")
predict = aa.predict(x_train)
print("Accuracy Score : ", round(accuracy_score(predict,y_train)*100,2),"%")
```

```python
print("Confusion Metrics \n",confusion_matrix(predict,y_train))
```

```python
print("Test prediction data report ")
predict = aa.predict(x_test)
print("Accuracy Score : ",round(accuracy_score(predict,y_test)*100,3),"%")
```
